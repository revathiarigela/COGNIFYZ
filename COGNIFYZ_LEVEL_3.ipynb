```python
# import python libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# import csv file
df = pd.read_csv('/content/Dataset.csv', encoding= 'unicode_escape')
```

```python
# Assuming your DataFrame is named 'df'

# Convert 'Has Table booking' column to a binary numerical feature
df['Has Table Booking'] = df['Has Table booking'].map({'Yes': 1, 'No': 0})

# Convert 'Has Online delivery' column to a binary numerical feature
df['Has Online Delivery'] = df['Has Online delivery'].map({'Yes': 1, 'No': 0})

# Display the updated DataFrame with the new binary numerical features
print(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Assuming your DataFrame is named 'df'

# Select the features (independent variables) and the target (dependent variable)
features = df[['Average Cost for two', 'Has Table Booking', 'Has Online Delivery', 'Votes']]
target = df['Aggregate rating']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Initialize the linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)

# Predict the target variable on the test data
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Assuming your DataFrame is named 'df'

# Select the features (independent variables) and the target (dependent variable)
features = df[['Average Cost for two', 'Has Table Booking', 'Has Online Delivery', 'Votes']]
target = df['Aggregate rating']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Initialize and train Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Predict the target variable using Linear Regression
y_pred_linear = linear_model.predict(X_test)

# Evaluate Linear Regression model
mse_linear = mean_squared_error(y_test, y_pred_linear)
r2_linear = r2_score(y_test, y_pred_linear)

print("Linear Regression - Mean Squared Error:", mse_linear)
print("Linear Regression - R-squared:", r2_linear)

# Initialize and train Decision Tree Regression model
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)

# Predict the target variable using Decision Tree Regression
y_pred_dt = dt_model.predict(X_test)

# Evaluate Decision Tree Regression model
mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)

print("Decision Tree Regression - Mean Squared Error:", mse_dt)
print("Decision Tree Regression - R-squared:", r2_dt)

# Initialize and train Random Forest Regression model
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)

# Predict the target variable using Random Forest Regression
y_pred_rf = rf_model.predict(X_test)

# Evaluate Random Forest Regression model
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("Random Forest Regression - Mean Squared Error:", mse_rf)
print("Random Forest Regression - R-squared:", r2_rf)
```

```python
# Create a histogram to visualize the distribution of ratings
plt.figure(figsize=(8, 6))
plt.hist(df['Aggregate rating'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()
```

```python
import seaborn as sns

# Create a bar plot to visualize the average rating for each price range
plt.figure(figsize=(10, 6))
sns.barplot(x='Price range', y='Aggregate rating', data=df, ci=None, palette='viridis')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.title('Average Rating by Price Range')
plt.show()
```

```python
# Create a box plot to visualize the distribution of ratings for each price range
plt.figure(figsize=(10, 6))
sns.boxplot(x='Price range', y='Aggregate rating', data=df, palette='Set2')
plt.xlabel('Price Range')
plt.ylabel('Aggregate Rating')
plt.title('Distribution of Ratings by Price Range')
plt.show()
```

```python
# Create a violin plot to visualize the distribution of ratings for each price range
plt.figure(figsize=(10, 6))
sns.violinplot(x='Price range', y='Aggregate rating', data=df, palette='coolwarm')
plt.xlabel('Price Range')
plt.ylabel('Aggregate Rating')
plt.title('Distribution of Ratings by Price Range')
plt.show()
```

```python
# Group the data by cuisine and calculate the average rating for each cuisine
average_rating_by_cuisine = df.groupby('Cuisines')['Aggregate rating'].mean().reset_index()

# Sort cuisines based on average rating in descending order
sorted_cuisines_by_rating = average_rating_by_cuisine.sort_values(by='Aggregate rating', ascending=False)

# Select the top N cuisines for visualization
top_n_cuisines = 10
top_cuisines = sorted_cuisines_by_rating.head(top_n_cuisines)

# Create a bar plot to compare the average ratings of top N cuisines
plt.figure(figsize=(12, 6))
sns.barplot(x='Cuisines', y='Aggregate rating', data=top_cuisines, palette='viridis')
plt.xlabel('Cuisines')
plt.ylabel('Average Rating')
plt.title(f'Top {top_n_cuisines} Cuisines with Highest Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Assuming your DataFrame is named 'df'

# Group the data by city and calculate the average rating for each city
average_rating_by_city = df.groupby('City')['Aggregate rating'].mean().reset_index()

# Sort cities based on average rating in descending order
sorted_cities_by_rating = average_rating_by_city.sort_values(by='Aggregate rating', ascending=False)

# Select the top N cities for visualization
top_n_cities = 10
top_cities = sorted_cities_by_rating.head(top_n_cities)

# Create a bar plot to compare the average ratings of top N cities
plt.figure(figsize=(12, 6))
sns.barplot(x='City', y='Aggregate rating', data=top_cities, palette='viridis')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.title(f'Top {top_n_cities} Cities with Highest Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
